# Question 1 : What is Simple Linear Regression (SLR)? Explain its purpose.

**Simple Linear Regression (SLR)** is a statistical and machine-learning technique used to model the relationship between one independent variable (X) and one dependent variable (Y) by fitting a straight line to observed data.

It is represented by the equation:

              Y = β_0 + β_1X + ε

where:


β_0 = intercept (value of Y when X = 0)

β_1 = slope (change in Y for a unit change in X)

ε = error term (difference between actual and predicted values)

**Purpose of Simple Linear Regression**

 -  To understand relationships – It helps identify and quantify how one variable affects another.

 -  To make predictions – It is used to predict the value of the dependent variable based on a given independent variable.

 -  To analyze trends – It reveals the direction (positive or negative) and strength of the relationship.

 - To support decision-making – Useful in fields like economics, business, and engineering for forecasting outcomes.

# Question 2: What are the key assumptions of Simple Linear Regression?

The **key assumptions of Simple Linear Regression (SLR)** are:

1. **Linearity**
   The relationship between the independent variable (X) and the dependent variable (Y) is linear.

2. **Independence of errors**
   The residuals (errors) are independent of each other (no autocorrelation).

3. **Homoscedasticity**
   The residuals have constant variance across all values of X.

4. **Normality of errors**
   The residuals are normally distributed (especially important for hypothesis testing).

5. **No extreme outliers**
   There are no significant outliers that unduly influence the regression line.


# Question 3: Write the mathematical equation for a simple linear regression model and explain each term.

The mathematical equation of a Simple Linear Regression model is:

            Y = β_0 + β_1X + ε

Explanation of each term:

Y: Dependent variable (response/output)
The variable we want to predict or explain.

X: Independent variable (predictor/input)
The variable used to predict the value of Y.

β_0(Intercept):
The expected value of Y when X=0. It represents where the regression line crosses the Y-axis.

β_1(Slope coefficient):
The change in Y for a one-unit increase in X. It shows the direction and strength of the relationship.

ε (Error term):
Represents random error or unexplained variation in Y not captured by the model.

# Question 5: What is the method of least squares in linear regression?

The method of least squares is a mathematical approach used in linear regression to estimate the best-fitting regression line for a given set of data points.

Explanation:

In Simple Linear Regression, the predicted values
Y^ are given by:

       Y^=β_0​+β_1​X

The least squares method chooses the values of β_0(intercept) and β_1(slope) such that the sum of the squares of the residuals is minimized.

        Minimize ∑(Y_i​−Y^_i​)2

where:

Y_i = actual observed value

Y^_i = predicted value

Y_i − Y^_i = residual (error)

**Purpose:**

-  Squaring the errors ensures all residuals are positive and penalizes large errors more heavily.

-  Minimizing this sum results in the most accurate and unbiased linear fit to the data.

**Difference between Logistic Regression and Linear Regression**

Linear Regression and Logistic Regression differ mainly in their purpose and output. Linear Regression is used to predict continuous numerical values, such as house prices or exam scores, and it models the relationship between variables using a straight-line equation. Its output can take any real value, and the model is typically trained using the method of least squares.

In contrast, Logistic Regression is used for classification problems, especially binary classification like pass/fail or yes/no outcomes. Instead of predicting exact values, it estimates the probability that an input belongs to a particular class. This probability is constrained between 0 and 1 using a sigmoid (logistic) function, and the model is trained using maximum likelihood estimation rather than least squares.

-  **Linear Regression** predicts numeric values.

-  **Logistic Regression** predicts class probabilities and is used for classification.

# Question 7: Name and briefly describe three common evaluation metrics for regression models.

Three common evaluation metrics for regression models are:

-  Mean Absolute Error (MAE):
It measures the average of the absolute differences between actual and predicted values. It shows how far predictions are from real values on average.

-  Mean Squared Error (MSE):
It calculates the average of the squared differences between actual and predicted values. Squaring penalizes larger errors more heavily.

-  Root Mean Squared Error (RMSE):
It is the square root of MSE and represents the average prediction error in the same units as the target variable, making it easier to interpret.

# Question 8: What is the purpose of the R-squared metric in regression analysis?

The purpose of the **R-squared (R²) metric** in regression analysis is to measure how well the regression model explains the variability of the dependent variable.

R-squared represents the proportion of total variance in the output variable that is explained by the independent variable(s). Its value ranges from 0 to 1, where:

-  0 means the model explains none of the variability,

-  1 means the model explains all the variability.

A higher R-squared value indicates a better fit of the model to the data, showing that the predictors effectively explain the outcome. However, it does not guarantee that the model is correct or suitable in all cases.

# Question 9: Write Python code to fit a simple linear regression model using scikit-learn and print the slope and intercept.
(Include your Python code and output in the code box below.)



```python
# Simple Linear Regression using scikit-learn

import numpy as np
from sklearn.linear_model import LinearRegression

# Sample data
X = np.array([1, 2, 3, 4, 5]).reshape(-1, 1)  # Independent variable
y = np.array([2, 4, 6, 8, 10])               # Dependent variable

# Create and train the model
model = LinearRegression()
model.fit(X, y)

# Print slope and intercept
print("Slope (Coefficient):", model.coef_[0])
print("Intercept:", model.intercept_)
```

# Question 10: How do you interpret the coefficients in a simple linear regression model?

In a simple linear regression model, the coefficients explain how the dependent variable changes with respect to the independent variable.

**Intercept (β_0):**
It represents the expected value of the dependent variable when the independent variable is zero. In practical terms, it is the starting point of the regression line on the Y-axis.

**Slope (β_1):**
It indicates the average change in the dependent variable for a one-unit increase in the independent variable.

 - A positive slope means the dependent variable increases as the independent variable increases.

 - A negative slope means the dependent variable decreases as the independent variable increases.

Overall, the coefficients quantify the direction, strength, and nature of the relationship between the two variables.
